```python
import pandas as pd
import numpy as ny
import os
```

```python
data = os.path.abspath(r"C:/Users/qiuhua.liu/")
os.chdir(data)
items = pd.read_csv('machine learning/train_monthly_nona.csv', header=0, sep=',',
                     quotechar='"')
items.head()
```

```python
items2=items.drop(['year','month','unit_sales','transactions','count_national_holiday','count_regional_holiday',
                 'count_local_holiday','count_days_onpromotion'],axis=1)
items2.head()
```

```python
items2['combination']=""
list_items=['item_nbr','store_nbr','store_city','store_state','store_type','store_cluster','item_family',
               'item_class','item_perishable']
for x in list_items:
    items2['combination']=items2['combination']+items2[x].astype(str)
items2.head()
 
```

```python
items3=items2.drop_duplicates(subset='combination', keep='first', inplace=False)
items3.head()
```
